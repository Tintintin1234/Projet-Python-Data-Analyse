```python
from sklearn.model_selection import train_test_split
#On créé 2 sets de données à partir de notre base de donnée initiale
def split(df):
    trainset, testset = train_test_split(df, test_size=0.3, random_state=0)
    return trainset,testset
```

```python
#On va remplacer les données qualitatives en données quantitatives pour mieux les integrer dans le modèle
données = {'no':1,
              'yes':1,
              'Female':0,
              'Male':1,
              'Sometimes':1,
              'Frequently':2,
              'Always':3,
              'Automobile':1,
              'Motorbike':2,
              'Bike':3, 
              'Public_Transportation':4,
              'Walking':5, 
              'Insufficient_Weight':0,
              'Normal_Weight':1 ,
              'Overweight_Level_I':2 ,
              'Overweight_Level_II':3,
              'Obesity_Type_I':4,            
              'Obesity_Type_II':5,
              'Obesity_Type_III':6
           }

def change(df):
    for col in df.select_dtypes('object').columns:
        df.loc[:,col] = df[col].map(données)
    
    return df
```

```python
change(dc)
print(dc)
```

```python
def preprocessing(df,name):
    df = change(df)
    X = df.drop(name,axis = 1)
    Y = df[name]
    
    return X, Y
```

```python
trainset, testset = split (dc)
X_test, Y_test = preprocessing(testset,'NObeyesdad')
X_train, Y_train = preprocessing(trainset,'NObeyesdad')
```

On va utiliser différents modèles et les comparer.

```python
#On utilise un modèle "decision tree"
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier
adaboost_model = AdaBoostClassifier(random_state=0)
model_decisiontree= DecisionTreeClassifier(random_state=0)
knn_model = KNeighborsClassifier()
boosting_model = GradientBoostingClassifier()
```

```python
#On importe les fonctions que l'on va utiliser depuis sklearn
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score, confusion_matrix, classification_report,make_scorer
from sklearn.model_selection import learning_curve
```

On va alors tester tous ces modèles et mesurer leur précision sur le même train et test set.

```python
def precision(model):
    model.fit(X_train, Y_train)
    ypred = model.predict(X_test)
    model_acc = accuracy_score(Y_test, ypred)
    print(confusion_matrix(Y_test, ypred)) 
    print(classification_report(Y_test, ypred))
    print('Précision : ' + str(model_acc))
    
print("Decision Tree")
precision(model_decisiontree)
print("")
print("Adaboost model")
precision(adaboost_model)
print("")
print("K plus proches voisins")
precision(knn_model)
print("")
print("Model Boosting")
precision(boosting_model)
```

Le modèle boosting est le plus précis (95.8%), suivit du modèle d'arbre de décision (93,8%). Le modèle des K plus proches voisins est lui aussi relativement précis mais moins que le précedent (86.9%) contrairement au modèle adaboost, qui lui n'est pas précis du tout (29,2%). 
Les performances du model adaboosting étant vraiment trop faible, on le passera pour la suite.

```python
print("1ere figure : Decision Tree")
pd.DataFrame(model_decisiontree.feature_importances_, index= X_train.columns).plot.bar(figsize=(12, 10))
print("2eme figure : Boosting Model")
pd.DataFrame(boosting_model.feature_importances_, index= X_train.columns).plot.bar(figsize=(12, 10))
```
